```python
class Vector:
    def __init__(self, vector):
        self.vector=vector
        
    def __add__(self, Vector_1):
        lista = []
        for i in range(len(Vector_1.vector)):
            lista.append(self.vector[i]+Vector_1.vector[i])
        return Vector(lista)
    
    def __len__(self):
        return(len(self.vector))
    
    def __repr__(self):
        return(str(self.vector))
    
    def __str__(self):
        return(str(self.vector))
        
    
```
